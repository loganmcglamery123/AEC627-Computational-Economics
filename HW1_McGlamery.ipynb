2.1  
The following function calculates $s(n)=\sum_{k=1}^n k^2$. The function takes $n$ as an input, and returns $s$ as the output.

```python
# 2.1
def sum_of_squares(n):
    s = sum(k**2 for k in range(1, n + 1))
    return s
```

2.3  
The following function solves $\max\{n\in \mathcal{N}: s(n)<K, K\in \mathcal{R}\}$, where $K$ is the inpurt argument of the function

```python
# 2.3
def max_n_for_sum(K):
    n = 0
    s = 0
    while s < K:
        n += 1
        s += n**2
    return n - 1  

#example usage
print("K=10, max n: ", max_n_for_sum(10))
print("K=100, max n: ", max_n_for_sum(100))
print("K=1000, max n: ", max_n_for_sum(1000))
```

3.1  
Plot of $sin(x)$ over $[0, 2*\pi]$  
alt text


```python
# 3.1
#  Define the range for x
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Define x-ticks in terms of multiples of pi
x_ticks = [0, np.pi / 2, np.pi, 3 * np.pi / 2, 2 * np.pi]
x_labels = ['0', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$']

# Plot sin(x) with x-axis increments in terms of pi
plt.figure(figsize=(8, 6))
plt.plot(x, y_sin, linestyle='-', marker='o', color='black', label='sin(x)')
plt.xticks(ticks=x_ticks, labels=x_labels)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Plot of sin(x)')
plt.grid(True)
plt.legend()
plt.show()
```

6. 

   OLS regression of variable pop_growth on med_hh_inc_2000 and bachelor2000:  

| Variable        |     Estimate |   t-value |    p-value |
|:----------------|-------------:|----------:|-----------:|
| Intercept       | -75.6016     | -2.85847  | 0.00428523 |
| med_hh_inc_2000 |   0.00177907 |  2.44545  | 0.0145224  |
| bachelor2000    |  -0.00937989 | -0.260613 | 0.794408   |

```python
# 6

import statsmodels.api as sm

# 1. Load the Excel dataset
file_path = "data_Dec2023.xlsx" 
data = pd.read_excel(file_path)

# 2. Define the dependent and independent variables
y = data["pop_growth"]  # Dependent variable
X = data[["med_hh_inc_2000", "bachelor2000"]]  # Independent variables

# 3. Add a constant term for the intercept
X = sm.add_constant(X)

# 4. Run the OLS regression
model = sm.OLS(y, X)
results = model.fit()

# 5. Report the estimates, t-values, and p-values
output = pd.DataFrame({
    "Variable": results.params.index,
    "Estimate": results.params.values,
    "t-value": results.tvalues.values,
    "p-value": results.pvalues.values
})

output["Variable"] = output["Variable"].replace({"const": "Intercept"})
# Display the results

markdown_table_5 = output.to_markdown(index=False)
print(markdown_table_5)
```
